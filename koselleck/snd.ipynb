```python
SND_GBY=['word','period']

def snd(word_or_words,gby=SND_GBY,num_proc=1,ntop=25):
    dfnbr=nbr(word_or_words,num_proc=num_proc)#.groupby(gby).mean()
    dfnbr['num_nbr']=1
    return dfnbr.query(f'rank<={ntop}').groupby(gby).agg(
        dict(num_nbr=np.sum, count=np.median, cdist=np.mean)
    )
```

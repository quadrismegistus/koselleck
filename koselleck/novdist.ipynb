```python
def respell_change(x):
    #x=x.replace('z','s')
    if x.endswith('.'): x=x[:-1]
    x=x.lower()
    if 'ific' in x: x=x.split('ific')[0]+'if'
    for y in ['izat',
              'isat',
              'icat',
              'ises',
              'icas',
              'ication',
#               'ation',
              'asion',
              'izes']:
        x=x.split(y)[0]
    if x.endswith('is') or x.endswith('iz'):
        x=x[:-2]
    if 'militaris' in x: x=x.replace('militaris','militar')
    if 'conceptual' in x: x=x.replace('conceptual','conc')
    if x=='abstraction': x='abstract.'
    if x.startswith('automob'): x='automob'
    x=x.replace('de-','de').replace('re-','re')
    x=dict(
        deverbal='nominal',
        reif='concret',
        financial='econom',
        embod='concret',
        petrol='automat',
        biol='biolog',
        desacral='demoral',
        disciplinar='professional',
        evaluation='reeval',
        defrancophon='transl',
        delatin='transl',
        revaluation='reeval',
        individual='personal',
        remoral='reval',
        deconcret='deconcret',
        dereif='deconcret',
        reval='reeval',
        automob='automat',
        spatial='respatial',
        geopolitic='politic',
        cheapen='declassing',
        tech='technolog',
        despatial='respatial',
        secular='secularis',
        dereligios='secularis',
        **{
            'global':'respatial',
            'globalis':'respatial'
        }
        
        
    ).get(x,x)
    
    if not x.endswith('.'): x=x+'.'
    return x
```

```python
def get_novdists(**opts):
    return get_changepoints(**opts)
    
def get_all_novdists(
        p_peak=1,
        nov_key='foote_novelty_perc',
        chpdist_key='chpdist_perc',
        selfdist_key='selfdist_perc',
        dist_key='dist_perc',
        no_fs=False,
        **opts):
    odf=get_changepoints()#p_peak=1,incl_words=True,**opts).dropna()
    if no_fs: odf=odf.loc[[w for w in odf.index if is_safe_fs(w)]]
    if p_peak and p_peak<1: odf=odf[odf.p_peak<=p_peak]
        
    s=odf['dist']=odf[[chpdist_key,selfdist_key]].mean(axis=1)
    odf['dist_rank']=s.rank(ascending=False)
    odf['dist_perc']=(s.rank(ascending=True) / len(s)) * 100
    
        
    s=odf['novdist']=odf[[nov_key,dist_key]].mean(axis=1)
    odf['novdist_rank']=s.rank(ascending=False)
    odf['novdist_perc']=(s.rank(ascending=True) / len(s)) * 100
    
    
    #odf['novdist_rank']=odf.novdist.rank(ascending=False)
    #odf['novdist_perc']=(odf.novdist.rank(ascending=True) / len(odf)) * 100
    odf['changepoint']=odf['chp']
    return odf.sort_values('novdist_rank')
```

```python

def get_signif_novdists(
        p_peak=.05,
        min_periods=1,
        min_nov=75,
        min_dist=75,
        min_selfdist=75,
        min_novdist=None,
        nov_key='foote_novelty_perc',
        dist_key='chpdist_perc',
        novdist_key='novdist_perc',
        selfdist_key='selfdist_perc',
        **opts):
    odf=get_all_novdists(
        p_peak=p_peak,
        min_periods=min_periods,
        nov_key=nov_key,
        dist_key=dist_key,
        **opts)
    
    if min_nov: odf=odf[odf[nov_key]>=min_nov]
    if min_dist: odf=odf[odf[dist_key]>=min_dist]
    if min_selfdist: odf=odf[odf[selfdist_key]>=min_selfdist]
    if min_novdist: odf=odf[odf[novdist_key]>=min_novdist]
    return odf.sort_values('novdist_rank')
```

```python
STZ_PRDS={'Stz.I','Stz.II'}

def in_hypersample(
        row,
        nov_key='foote_novelty_perc',
        chpdist_key='chpdist_perc',
        selfdist_key='selfdist_perc',
        dist_key='dist_perc',
        thresh=85):
    if not row.in_sample: return False
    if not row[nov_key] or row[nov_key]<=thresh: return False
    if not row[chpdist_key] or row[chpdist_key]<=thresh: return False
    if not row[selfdist_key] or row[selfdist_key]<=thresh: return False
    if not row[dist_key] or row[dist_key]<=thresh: return False
    return True

def get_tagsample(df,
                 period_len=20,
                 n=10,
                 ):
    df['in_tagsample']=False
    df.loc[df.in_hypersample==True, 'in_tagsample']=True
    df.loc[((df.is_ever_abs==True) & (df.in_sample==True)), 'in_tagsample']=True
    
    num_abs_so_far=len(df.query('in_tagsample==True & is_ever_abs==True'))
    num_notabs_so_far=len(df.query('in_tagsample==True & is_ever_abs==False'))
    num_notabs_needed = num_notabs_so_far - num_abs_so_far
    
    df['period2']=df['period'].apply(lambda y: force_year(y)//period_len*period_len)
    
    
    dfsmpl_q=df[~df.period2.isna()].query('in_sample==True & is_ever_abs==False & is_valid_bool=="Unknown"')
    dfsmpl_q.period2=dfsmpl_q.period2.apply(int)
    dfsmpl_g=dfsmpl_q.sort_values('novdist_rank').groupby(['period2','pos'])
    n=num_notabs_needed//len(dfsmpl_g) if not n else n
    dfsmpl=dfsmpl_g.head(n=n)
#     display(dfsmpl.groupby(['period2','pos']).size())
#     display(dfsmpl.sort_values('period2'))
    
    df.loc[dfsmpl.index, 'in_tagsample']=True
    
    return df

def get_minisample(df,n=10,only_abs=True,abs_cutoff=0):
    #dfsmpl=df.query('in_tagsample==True & is_valid_bool!="Invalid"').sort_values('novdist_rank')
    dfq=df.query(f'in_sample==True')
    if only_abs:
        absw=get_words_ever_abs(cutoff=abs_cutoff)
        dfq=dfq.loc[set(dfq.index) & set(absw)]
    dfsmpl=dfq.groupby(['period2','pos']).head(n=n)
    df['in_minisample']=False
    df.loc[dfsmpl.index, 'in_minisample']=True
    return df

def get_all_novdists_anno(force=False,only_stz=False,only_pos=set(),**signif_opts):
    # merge
    dfall=get_all_novdists().assign(in_sample=False)
    dfsignif=get_signif_novdists(**signif_opts).assign(in_sample=True)
    df=dfsignif.append(dfall).reset_index().drop_duplicates('word').set_index('word')
    df['period_stz']=df['chp'].apply(periodise_stz_nov)
    if only_stz: df=df[df.period_stz.isin(STZ_PRDS)]
    if only_pos: df=df[df.pos.isin(only_pos)]
    
    # merge with anno
    dfanno=get_dfanno(force=force)
    dfanno=dfanno[[c for c in dfanno.columns if c not in set(df.columns)]]
    df=dfanno.join(df,how='outer').fillna('')
    
    # add etc
    absw=get_words_ever_abs(cutoff=1)
    df['is_ever_abs']=[w in absw for w in df.index]
    df['random']=[random.random() for n in range(len(df))]
    df['in_hypersample']=df.progress_apply(in_hypersample,axis=1)
    df=df.sort_values('random')
    df=get_tagsample(df)
    df=get_minisample(df)
    
    cols = ANNOCOLS + ['neighb1','neighb2','changepoint','chpdist_perc','selfdist_perc','novdist_perc','foote_novelty_perc']
    df=df[[c for c in cols if c in set(df.columns)] + [c for c in sorted(df.columns) if c not in set(cols)]]
    for bc in ['','_period']:
        if bc in set(df.columns):
            df=df.drop(bc,1)
    return df.sort_values(['period2','pos'])
    
    
```

```python
def show_changepoints_anno(
        df=None,eg_lim=100,period_len=20,#no_fs=False,
        only_valid=False,in_sample=True,lim_minieg=1,
        only_abs=False,
        pos_3way=False,
        abs_cutoff=1,
        in_hypersample=False,
        in_minisample=False,
        in_tagsample=True,
        **opts):
    odf=get_all_novdists_anno(
        **opts,
    ).query('novdist!=""').sort_values('novdist').dropna().reset_index()
    odf['change']=odf.Changes.fillna('').apply(
        lambda x: ''# if not to_words(x) else respell_change(to_words(x)[0]).title()
    )
    invalid=get_invalid_words()
    odf=odf[~odf.word.isin(invalid)]
    if only_valid: odf=odf[odf.is_valid.str.startswith('y')]
    if in_sample: odf=odf[odf.in_sample==True]
    if in_hypersample: odf=odf[odf.in_hypersample==True]
    if in_minisample: odf=odf[odf.in_minisample==True]
    if in_tagsample: odf=odf[odf.in_tagsample==True]
    abswords=get_words_ever_abs(cutoff=abs_cutoff)
    if only_abs: odf=odf[odf.word.isin(abswords)]
    o=[]
    def row2str(row):
        ostr=o=f'{row.word}'
        if row.From.strip() and row.To.strip():
            a=to_words(row.From)
            b=to_words(row.To)
        else:
            #a=[w[1:] for w in row.neighb1[1:-1].split() if w.startswith('-') and w[1:]!=row.word]
            #b=[w[1:] for w in row.neighb2[1:-1].split() if w.startswith('+') and w[1:]!=row.word]
            #if not a: a=to_words(row.neighb1)
            #if not b: b=to_words(row.neighb2)
            a=to_words(row.neighb1)
            b=to_words(row.neighb2)
            
        astr='/'.join(a[:lim_minieg])
        bstr='/'.join(b[:lim_minieg])
        ostr+=f' <small>({astr}â†’{bstr})</small>'
        if row.is_valid.lower().strip().startswith('y'):
            ostr=f'<i>{ostr}</i>'
        return ostr
        
    
    odf['period']=odf['period'].apply(lambda y: force_year(y)//period_len*period_len)
    if pos_3way: odf['pos']=odf['pos'].apply(lambda x: str(x)[0])
    for period,g in tqdm(odf.groupby('period')):
        periodstr=f'{period:.0f}-{period+period_len:.0f}'
        od=dict(period=periodstr)
        
        for pos,gpos in g.groupby('pos'):
            wordstrs=[]
            for ch,chgpos in sorted(
                    gpos.groupby('change'),
                    key=lambda _g: (int(_g[0]==''),-len(_g[1]))
                    ):
                chgpos=chgpos.sort_values('novdist_rank')
                wordstrx=f'<i>{ch}</i>: ' if ch else ''
                chgpos_strs=chgpos.iloc[:eg_lim].apply(row2str,axis=1)
#                 wordstrx+=', '.join(chgpos_strs)                
                wordstrx+='<li>' + '</li><li>'.join(chgpos_strs) + '</li>'
                wordstrs+=[wordstrx]
            #wordstr='<li>' + '</li><li>'.join(wordstrs) + '</li>'
            wordstr='<hr/>'.join(wordstrs)
            od[pos]=wordstr
        o.append(od)
    odf=pd.DataFrame(o).set_index('period').sort_index().fillna('')
    
    POSD['n']='Nouns'
    if not pos_3way:
        poscols=['nn1','nn2','j','v']
    else:
        poscols=['n','j','v']
    poscols=[x for x in poscols if x in set(odf.columns)]
    odf=odf[poscols]
    odf.columns=[POSD.get(x,x) for x in poscols]
    
    return odf
        
```

```python
def make_figdf(dfsign,dtype='signif_word',
               rolling_n=1,rolling_min_periods=1,
               min_all_periods=None,only_max_periods=True,
               ymin=YMIN_DISTMAT,ymax=YMAX_DISTMAT):
    l=[]
    idf=dfsign
    for pos,dfsign_pos in idf.groupby('pos'):
        odf=dfsign_pos.set_index('period').sort_index()
        odf['num_peaks']=odf.groupby('period').size()
        odf['num_peaks_rel']=odf.num_peaks / odf.num_peaks.max() * 100
        if rolling_n and rolling_n>1:
            odf=odf.rolling(rolling_n,min_periods=rolling_min_periods).mean()
        odf['dtype']=dtype
        odf['pos']=pos
        l+=[odf.reset_index()]
    odf=pd.concat(l) if len(l) else pd.DataFrame()
    odf=odf.groupby(['period','foote_size','pos']).mean().reset_index()
    return odf

def is_safe_fs(w):
    return not ('s' in w or 'f' in w)
```

```python
def plot_novdist_signif(figdf=None,y='num_peaks',save=False,showdata=False,vnum=gettodaysdate(),
                    key='_figdf_nov_signif_',force=False,
                    rolling_n=1,rolling_min_periods=1,returndata_all=False,force_data=False,
                    by_foote_size=False,min_dist_perc=None,
                    ymin=1755,ymax=1925,nudge_x=2,
                    labsize=6
                   ):
    p9.options.dpi=300
    
    
    defaults=dict(
        min_foote_size=FOOTE_W,max_foote_size=FOOTE_W,
        p_peak=.01,consecutive_peaks=True,min_periods=1,
        ymin=YMIN_DISTMAT,ymax=YMAX_DISTMAT,ybin=YBIN_DISTMAT,
        force=force_data,only_peak_scores=True,
        min_selfdist=None,min_dist=None,min_nov=None,
        only_first_peak=False,
        no_fs=False,
        #rolling_n=rolling_n,rolling_min_periods=rolling_min_periods,
    )
    
    if figdf is None and not returndata_all:
        if not force:
            with get_db('novdist') as db:
                if key in db: figdf=db[key]
    
    if figdf is None or returndata_all:
        cat_dfs=[
            (
                ('I','(a) All words'),
                #get_signif_novdists(**{**defaults, 'p_peak':.05})
                get_all_novdists(),
            ),
            (
                ('II','(b) Significantly novel (p<.05)'),
                get_signif_novdists(**{**defaults, 'p_peak':.05})
            ),
            (
                ('III','(c) High change magnitude (>75%)'),
                get_signif_novdists(**{**defaults, 'p_peak':.05, 'min_dist':75, 'min_selfdist':75,})
            ),
            (
                ('IV','(d) Without "f" or "s"'),
                get_signif_novdists(**{**defaults, 'p_peak':.05, 'min_dist':75, 
                                       'min_selfdist':75,'no_fs':True})
            ),
            
#             (
#                 ('IV','IV. Words with a significant novelty score (p<.01) and change (>%90)'),
#                 get_signif_novdists(**{**defaults, 'p_peak':.01, 'min_dist':99})
#             ),
            
        ]
        cats,dfl=zip(*cat_dfs)
        if returndata_all:
            return pd.concat(
                dfx.assign(dtype2=cat[1],dtype3=cat[1])
                for cat,dfx in zip(cats,dfl)
            )
        l=[]
        for cat,dfx in zip(cats,dfl):
            odfx=make_figdf(
                dfx,rolling_n=rolling_n,rolling_min_periods=rolling_min_periods
            ).assign(dtype2=cat[1], dtype3=cat[0])
            l+=[odfx]
        
            
        figdf=pd.concat(l)
        figdf['dtype2']=pd.Categorical(figdf.dtype2, categories=[yy for xx,yy in cats])
        figdf['dtype3']=pd.Categorical(figdf.dtype3, categories=[xx for xx,yy in cats])
        with get_db('novdist','w') as db: db[key]=figdf
    
    if ymin: figdf=figdf[figdf.period>=ymin]
    if ymax: figdf=figdf[figdf.period<ymax]
        
    figdf=pd.concat(
        g.assign(num_peaks_rel=g.num_peaks / g.num_peaks.max() * 100)
        for i,g in figdf.groupby(['pos','dtype2'])
    )
        
    figdf['pos']=pd.Categorical(
        figdf.pos.apply(lambda x: POSD.get(x,x)),
        categories=[POSD.get(x,x) for x in ['nn1','nn2','j','v']]
    )
    figdf=figdf.groupby(['pos','dtype2']).filter(lambda _g: len(_g)>=10)
    figdf.num_peaks=figdf.num_peaks.apply(int)
    if showdata: display(figdf)
    fig=start_fig(
        figdf,
        x='period',
        y=y,
        shape='dtype2',
        color='dtype2',
        theme='classic',
#         size='foote_novelty',
        label='num_peaks',
        figure_size=(8,8)
    ) + p9.geom_line(size=.666) + p9.geom_point(alpha=.5)
    label_periods(fig)
    #fig+=p9.facet_wrap('dtype2',ncol=1,scales='free_y')
    fig+=p9.facet_wrap('pos',ncol=1,scales='free_y')
    fig+=p9.geom_text(size=5,color='black')
    fig+=p9.scale_color_gray(start=.25,end=.75)
    fig+=p9.scale_size_continuous([.1,3])
    fig+=p9.labs(
        title='Number of words which "pivot", or reach their peak novelty score, in a given period',
        y='Number of word pivots (% of maximum)',
        x='Pivot period for word (peak novelty)',
        size='Foote novelty score (avg.)',
        color='Significance threshold',
        shape='Significance threshold'
    )
    #fig+=p9.theme(subplots_adjust={'bottom': 0.2})
    fig+=p9.theme(
        legend_position=(.5, 0.025),
        legend_direction='horizontal',
        legend_title=p9.element_blank()
    )
    fig+=p9.scale_x_continuous(breaks=list(range(1740,1940,20)))
    if save:
        ofn=f'fig.nov_summary.{y}.{vnum}.png'
        fig.save(os.path.join(PATH_FIGS,ofn))
        shutil.copyfile(os.path.join(PATH_FIGS,ofn), os.path.join(PATH_FIGS2,ofn))
    
    return fig
```

```python
def get_anno_graph(
        period_col='period_stz',
        anno_col='Changes',
        ybin=5,
        size_by='degree_centrality',
        size_min=10,
        num_modules=None,
        incl_words=True,
        ymin=1700,
        directed=False,
        p_peak=.01,
        min_dist=90,
        only_stz=True,
        min_ch_num=2,
        only_pos=None,
        only_valid=True,
        in_sample=True,
        **novdist_anno_opts
        ):
    g=nx.Graph() if not directed else nx.DiGraph()
    
    df=get_all_novdists_anno(p_peak=p_peak,min_dist=min_dist,only_stz=only_stz_num,
                             only_valid=True,in_sample=True,only_pos=only_pos,**novdist_anno_opts)
    
    
    for word,row in df.iterrows():
        prd=str(row[period_col])
        prd0=str(row['period_stz'])
        
        # add period
        if not g.has_node(prd): g.add_node(prd,node_type='Period',period=prd)
            
        if only_stz:
            if not g.has_node(prd0): g.add_node(prd0,node_type='Stz',period=prd0)
            if prd!=prd0: g.add_edge(prd0,prd,weight=1)
        
        if incl_words:
            if not g.has_node(word):
                g.add_node(word,node_type=f'Word',period=prd0)
            g.add_edge(prd,word,weight=1)
        
        
        anno=row[anno_col]
        anno_l=to_words(anno)
        for annox in anno_l:
            annox=respell_change(annox).title()
            if not g.has_node(annox):
                g.add_node(annox, node_type='Change')
            if incl_words:
                g.add_edge(word,annox,weight=1)
            else:
                if not g.has_edge(prd,annox):
                    g.add_edge(prd,annox,weight=1)
                else:
                    g[prd][annox]['weight']+=1
    
    
    #all_ch=Counter(n for n,d in g.nodes(data=True) if d['node_type']=='Change')
    if min_ch_num>1:
        all_ch=g.degree()

        for n,d in list(g.nodes(data=True)):
            if d['node_type']=='Change' and all_ch[n]<min_ch_num:
                g.remove_node(n)

    
    
    fl=['degree','degree_centrality','eigenvector_centrality','betweenness_centrality','eccentricity']
    for fn in fl:
        f=getattr(nx,fn)
        for n,v in list(dict(f(g)).items()):
            g.nodes[n][fn]=v
    
    nodes=list(g.nodes(data=True))
    data=[g.nodes[n][size_by] for n,dat in nodes]
    data_perc=[percentileofscore(data,x) for x in data]
    
    for (n,dat),perc in zip(nodes,data_perc):
        dat['neighbors']=', '.join(str(x) for x in g.neighbors(n))
        nt=dat['node_type']
        if nt.startswith('Word'):
            dat['size']=1
        else:
            #dat['size']=np.log(dat['degree'])
            dat['size']=dat['degree_centrality']*10
        
        if nt=='Stz':
            dat['size']+=20
        elif nt=='Period':
            dat['size']+=25
        elif nt=='Change':
            dat['size']+=15
        else:
            dat['size']+=10
            
    giant = max(nx.connected_components(g), key=len)
    [g.remove_node(n) for n in set(g.nodes()) - giant]
    
    try:
        if not num_modules: num_modules=len([n for n,d in g.nodes(data=True) if d['node_type']=='Period'])
        for i,grp in enumerate(tqdm(community.asyn_fluidc(g,num_modules))):
    #     for i,grp in enumerate(tqdm(community.greedy_modularity_communities(g))):
            for n in grp:
                g.nodes[n]['module']=i
    except Exception as e:
#         pass
        print("!!",e)
    
    badnodes=['Stz.NA']
    for bn in badnodes:
        if g.has_node(bn): g.remove_node(bn)
            
    periods=[n for n,d in g.nodes(data=True) if d['node_type']=='Period']        
    for n,d in g.nodes(data=True):
        if not d.get('period'):
            period2len=dict(
                (prd,nx.shortest_path_length(g,n,prd))
                for prd in periods
            )
            s=pd.Series(period2len).rank().sort_values()
            if s.iloc[0]==1:
                period=s.index[0]
                g.nodes[n]['period']=period
    return g
            
```

```python
def plot_anno_graph(g,forcefield=3.0,pos=None,layout_forceatlas=True,force_layout=False,layout_forceatlas_iter=2000,color_by='node_type',return_pos=False):
    if pos is None or force_layout:
        if layout_forceatlas:
            forceatlas2 = ForceAtlas2(
                # Behavior alternatives
                outboundAttractionDistribution=False,  # Dissuade hubs
                edgeWeightInfluence=1.0,
                jitterTolerance=1.0,  # Tolerance
                barnesHutOptimize=True,
                barnesHutTheta=1.2,
                scalingRatio=forcefield,
                strongGravityMode=False,
                gravity=1.0,
                verbose=False#True
            )

            pos = forceatlas2.forceatlas2_networkx_layout(
                g,
                pos=pos,
                iterations=layout_forceatlas_iter
            )
    if return_pos: return pos
            
            
    hvg = hv.Graph.from_networkx(
        g,
        nx.layout.spring_layout,
        weight='weight',
        iterations=0 if pos is not None else 50,
        pos=pos
    ).opts(
        node_color=hv.dim(color_by),
        edge_line_width=hv.dim('weight'),
        node_size=hv.dim('size'),
#         node_size=np.log10(hv.dim('degree')) * 25,
#         cmap=['#CCCCCC','#AAAAAA','#EEEEEE'],
        cmap='Set2',
        node_fill_alpha=1,
        node_line_alpha=0,
        edge_line_alpha=.25,
        inspection_policy='nodes',
        tools=[]#'pan','wheel_zoom','save']
    )

    labels = hv.Labels(
        hvg.nodes,
        ['x','y'],
        'index',
    ).opts(
        text_font_size='8pt',
#         text_font_style='bold',
        text_color='black',
    )
    return hvg*labels
```

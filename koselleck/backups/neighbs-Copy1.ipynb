```python
def do_nbr_word(dfgrp,max_rank=100,num_proc=1,progress=True):
    for word,period,run in dfgrp.groupby(dfgrp.index):
        print(word,period,run)
        return pd.DataFrame()
    
    row=dfgrp.reset_index().iloc[0]
    word,period,run=row.word,row.period,row.run
    
    
    dfprd=dfcdist.groupby('run').head(max_rank)
    dfprdg=dfprd.groupby(gby)
    dfprd=dfprd.reset_index().set_index(gby)
    dfprd['count']=dfprdg.size()
    dfprd['score']=[(c - (cd/10)) for c,cd in zip(dfprd['count'], dfprd['cdist'])]
    odf=dfprd.groupby(gby).agg(dict(
        count=np.median,
        score=np.mean,
        run=np.max,
        cdist=np.mean
    ))
    odf['rank']=odf['score'].rank(ascending=False,method='min').apply(int)
    odf=odf.drop('score',1).sort_values('rank')
    odf=odf.rename({'run':'num_runs'},axis=1)
    if max_rank: odf=odf[odf['rank']<=max_rank]
#     if cache:
#         with get_veclib('nbr',autocommit=True) as vl:
#             vl[qstr]=odf
    return odf if not cache_only else pd.DataFrame()



def nbr_word(word,period_or_periods=None,run_or_runs=None,neighbors=None,num_runs=10,num_proc=1,progress=True,
             max_rank=100):
    if progress: print(f'Running nbr_word({word})')
    dfc = cdist_word(word,period_or_periods,neighbors=neighbors,num_proc=num_proc,progress=progress)
    odf=pmap_groups(
        do_nbr_word,
        dfc.groupby(dfc.index),
        kwargs=dict(max_rank=max_rank,num_proc=1,progress=False),
        num_proc=num_proc,
        progress=progress
    )
    if progress: print(f'Finished running nbr_word({word})')
    return odf
```

### Scaling up

```python
def _to_nbr(objd): return to_nbr(**objd)
def _nbr_word(objd): return nbr_word(**objd)
def nbr_word(word,periods=None,ymin=YMIN,ymax=YMAX,ybin=YEARBIN,
            num_proc=1,progress=True,cache=True,cache_only=False,force=False):
    if not periods: periods=get_default_periods(ymin=ymin,ymax=ymax,ybin=ybin)
    wqstr=f'{word},ymin={YMIN},ymax={YMAX},ybin={YEARBIN}'
    if cache and not force:
        with get_veclib('nbr',autocommit=False) as vl:
            if wqstr in vl: return vl[wqstr]
            
    dfcdist=cdist(word,num_proc=num_proc,progress=progress)
    if not len(dfcdist): return pd.DataFrame()
    
    objs=[
        {
             'word': word,
             'period': period,
             'max_rank': 100,
             'force': False,
             'cache_only': False,
             'num_proc': 1,
             'progress': False,
             'cache_cdist': False,
             'cache': False,
             'dfcdist':prddf
        }
        for period,prddf in dfcdist.groupby('period')
        if period in set(periods)
    ]
    o=pmap(
        _to_nbr,
        objs,
        num_proc=num_proc,
        desc='Getting neighbors across periods',
        progress=progress,
    )
    odf=pd.concat(o) if len(o) else pd.DataFrame()
    if cache:
        with get_veclib('nbr',autocommit=True) as vl:
            vl[wqstr]=odf
    
    return odf if not cache_only else pd.DataFrame()
```

```python

def nbr(word_or_words,period_or_periods=None,prefix='nbr',neighbors=None,cache=True,
        max_rank=100,force=False,cache_only=False,num_proc=1,cache_cdist=False,progress=True):
    # preproc input
    words=tokenize_fast(word_or_words) if type(word_or_words)==str else list(word_or_words)
    if period_or_periods is None:
        periods=get_default_periods()
    elif type(period_or_periods)==str:
        periods=tokenize_fast(period_or_periods)
    else:
        periods=list(period_or_periods)
    objs=[
        dict(
            word=word,
            periods=periods,
            num_proc=1 if len(words)>1 else num_proc,
            progress=len(words)==1,
            cache=cache,
            cache_only=cache_only
        ) for word in words
    ]
    o=pmap(_nbr_word, objs, num_proc=num_proc,
           desc='Gathering neighborhoods across words',
           progress=progress if len(words)!=1 else False)
    return pd.concat(o) if len(o) else pd.DataFrame()

def nbr_(argd):
    try:
        return to_nbr(**argd)
    except Exception as e:
        pass
    return pd.DataFrame()
```

## Importing older data

```python
def load_older_data_neighbs(ifn=FN_ALL_NEIGHBS):
    df=read_df(ifn)
    df=df.query(f'"{YMIN}"<=period<"{YMAX}"')
    df['cdist']=1-df['csim']
    df['num_runs']=10
    df['count']=df['count'].apply(int)
    odf=df.set_index(['word','neighbor','period'])[['count','num_runs','cdist','rank']]
    return odf
```

```python
dfold=load_older_data_neighbs()
dfold
```

```python
def import_older_data_neighbs(df):
    with get_veclib('nbr',autocommit=False) as vl:
        for i,(word,wdf) in enumerate(tqdm(df.groupby('word'))):
            wqstr=f'{word},ymin={YMIN},ymax={YMAX},ybin={YEARBIN}'
            if not wqstr in vl:
                vl[wqstr]=wdf
            if i and not i%100: vl.commit()
        vl.commit()
```

```python
import_older_data_neighbs(dfold)
```

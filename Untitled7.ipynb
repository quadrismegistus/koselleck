```python
dfall=pd.read_pickle(FN_VECTOR_SCORES_RUNS)
dfall
```

```python
ever_abs=set(get_words_ever_abs())
dfall['ever_abs']=['Once abstract' if w in ever_abs else 'Never abstract'
                   for w in dfall.word]
```

```python
# df=dfall.groupby(['period','word']).mean()
# gby=['ever_abs','period','word']
gby=['ever_abs','period']#,'run']
df=dfall.groupby(gby).mean()
df=df.reset_index().melt(id_vars=gby,var_name='vector')
df
```

```python
gby2=['period']#,'run']
df2=dfall.groupby(gby2).mean()
df2=df2.reset_index().melt(id_vars=gby2,var_name='vector')
df2['ever_abs']='(All words)'
df2
```

```python
figdf=df#.append(df2)
figdf['year']=[int(x[:4]) for x in figdf.period]
figdf
```

```python
fig=start_fig(
    figdf,
    x='year',
    y='value',
    color='vector',
    group='vector',
    figure_size=(9,7)
) + p9.geom_point(size=.2) + p9.geom_smooth(se=True,method='loess') + p9.facet_wrap('ever_abs',nrow=1)
fig
```

```python
fig=start_fig(
    figdf.query('ever_abs=="Once abstract"'),
    x='year',
    y='value',
    color='vector',
    group='vector',
    figure_size=(9,7)
) + p9.geom_point(size=.2) + p9.geom_smooth(se=True,method='loess') + p9.facet_wrap('vector',scales='free_y')#,nrow=1)
fig
```

```python
figdf2=figdf[figdf.vector.isin({'Ambig.NX','Sing-Plural.M','Freq.M','Abs-Conc.Median'})]
figdf2['vector']=figdf2.vector.replace(
    {
        'Ambig.NX': 'Ambiguity',
        'Sing-Plural.M': 'Singularity',
        'Freq.M':'Frequency',
        'Abs-Conc.Median':'Abstractness'
    }
)
figdf2
```

```python
politvec={
#     'Collective-Indiv.RH':'Collectivization',
    'Woman-Man.VG':'Genderization',
    'Time-Space.HGI':'Temporalization',
    'Polit-Acad.HGI':'Politicization',
    'Interp-Desc.HGI':'Ideologization'
}
figdf3=figdf[figdf.vector.isin(set(politvec.keys()))]
figdf3['vector']=figdf3.vector.replace(politvec)
figdf3
```

```python
fig=start_fig(
    figdf3.query('1700<=year<1900'),
    x='year',
    y='value',
#     color='ever_abs',
    group='ever_abs',
    shape='ever_abs',
    figure_size=(8,6)
) + p9.geom_point(size=2) + p9.geom_smooth(se=True,method='loess',alpha=.5) + p9.facet_wrap('vector',scales='free_y')
fig+=p9.scale_color_gray()
ofnfn='figures/fig.fourway-vec-cmp.v2.vecpolit.png'
fig+=p9.labs(x='Date of semantic model', y='Standardized score',
            title='Abstract and concrete words along four semantic measurements, 1700-1900')
fig.save(ofnfn)
upfig(ofnfn)
fig
```

```python
def summarize_classed_changes(dfcc=None):
    if dfcc is None: dfcc=get_classed_changes()
    d={}
    for gi,gdf in dfcc.groupby('word'):
        d[gi]=' '.join([
#             ''.join(y for y in x if not y.isalpha()) + ''.join(y for y in x if y.isalpha())
            ''.join(y for y in x if y.isalpha()) + ''.join(y for y in x if not y.isalpha())
            for x in gdf.change if x and len(x)>1
            and x not in {'-Changed', 'Changed-'}
        ])
        
    return pd.Series(d).fillna('')
```

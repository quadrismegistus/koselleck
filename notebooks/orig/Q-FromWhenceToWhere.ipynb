```python
def classify_vector_means(row,perc_threshold=75,zcut=1):
    name=row.vector
    ns=str(name).split('.')[0].split('-')
    n1,n2=ns[0],ns[-1]
    o1,o2=None,None
    mean1,mean2=row.mean1,row.mean2
    if mean1>zcut: o1=n1
    if mean1<-zcut: o1=n2
    if mean2>zcut: o2=n1
    if mean2<-zcut: o2=n2
    return o1,o2#(o1,o2) if o1!=o2 else (None,None)

df['class1'],df['class2'] = zip(*df.apply(classify_vector_means,axis=1))
```

```python
def make_sankey_data(df):
    ld=[]
    for w,wdf in df.groupby('word'):
        classes1=wdf.class1.unique()
        classes2=wdf.class2.unique()
        shared=set(classes1) & set(classes2)
        
        classes1=[c for c in classes1 if c and c not in shared]
        classes2=[c for c in classes2 if c and c not in shared]
        if w=='peace':
            print(classes1)
            print(classes2)
            display(wdf)
        
        for c1 in classes1:
            for c2 in classes2:
                ld+=[{'word':w, 'start':c1, 'end':c2}]
    return pd.DataFrame(ld)
```

```python
dfsankey=make_sankey_data(df)
dfsankey
```

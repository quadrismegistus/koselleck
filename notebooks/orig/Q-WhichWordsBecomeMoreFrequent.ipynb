# Frequency

```python
from koselleck import *
```

```python
C=lltk.load('bpo')
```

```python
words = set(get_dfstat().index)
len(words),random.sample(words,5)
```

```python
dtm=C.dtm(words=words,tf=False)
dtm=dtm[[c for c in dtm.columns if not c in {'amp'}]]
dtm
```

```python
mdtm=C.meta[['year']].join(dtm,rsuffix='_w')
mdtm
```

```python
mdtmy=mdtm.groupby('year').sum()
mdtmy
```

```python
mdtmys=mdtmy.apply(lambda x: x/x.sum(),1)
mdtmys
```

```python
mdtmys.loc[1681]
```

```python
def plot_word(w):
    import plotnine as p9
    figdf=mdtmys.reset_index()[['year',w]].query('1680<=year<1940')
    figdf[w]*=1000000
    figdf=figdf.set_index('year').rolling(10).mean().reset_index()
    fig=p9.ggplot(figdf, p9.aes(x='year',y=w))
    fig+=p9.geom_point(size=0.1)
    fig+=p9.geom_smooth()
    return fig
```

```python
plot_word('culture')
```

```python
def plot_words(words,z=False):
    import plotnine as p9
    figdf=mdtmys.reset_index()[['year']+words].query('1680<=year<1940')
    figdf=figdf.set_index('year')
    figdf=figdf.rolling(10).mean()
    figdf*=1000000
    if z: figdf=figdf.apply(lambda x: (x-x.mean())/x.std())
    figdf=figdf.reset_index().melt(id_vars=['year'],value_vars=words,var_name='word',value_name='freq')
    fig=p9.ggplot(figdf, p9.aes(x='year',y='freq',color='word'))
    fig+=p9.geom_point(size=0.1)
    fig+=p9.geom_smooth()
    return fig
```

```python
plot_words(['culture','industry','value','liberty'],z=False)
```

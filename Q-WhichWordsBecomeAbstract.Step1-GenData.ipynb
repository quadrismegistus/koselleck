```python
dfpiv_abs_z=get_dfpiv_abs(words,z=True)
dfpiv_abs_z
```

```python
def compute_means(w,df):    
    grp1=df[[c for c in df.columns if c<1770]].loc[w]
    grp2=df[[c for c in df.columns if c>=1830]].loc[w]
    mean1=grp1.mean()
    mean2=grp2.mean()
    t,p=ttest_ind(grp1.dropna().values, grp2.dropna().values)
    return {'word':w, 'mean1':mean1, 'mean2':mean2, 'mean_diff':mean2-mean1, 'mean_diff_t':t, 'mean_diff_p':p}
```

```python
compute_means('elbow', dfpiv_abs_z)
```

```python
compute_means('station', dfpiv_abs_z)
```

```python
compute_means('history', dfpiv_freq_z)
```

```python
def compute_means_df(dfpiv):
    odf=pd.DataFrame([compute_means(w,dfpiv) for w in tqdm(dfpiv.index,position=0)])
    odf=odf.dropna().sort_values('mean_diff').set_index('word')
    odf['mean_diff_magn']=odf.mean_diff.apply(abs)
    
    for x in ['mean_diff','mean_diff_magn']:
        odf[f'{x}_rank']=(odf[x]*-1 if x.endswith('_magn') else odf[x]).rank()
        odf[f'{x}_z']=(odf[x] - odf[x].mean())/odf[x].std()
        odf['mean_diff_magn_rank']=(odf.mean_diff_magn*-1).rank()
    return odf
```

```python
meandiff_abs = compute_means_df(dfpiv_abs_z)
meandiff_abs
```

```python
meandiff_freq = compute_means_df(dfpiv_freq_z)
meandiff_freq
```

```python
meandiff = meandiff_abs.join(meandiff_freq,how='inner',lsuffix='_abs',rsuffix='_freq')
meandiff
```

```python
meandiff.to_csv(OFN_MEAN_DIFF)
```

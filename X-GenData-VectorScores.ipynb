```python

def get_misccontrasts(fieldd=None):
    if fieldd is None: fieldd=get_fields()
    return [{
        'contrast':'Woman-Man',
        'source':'S',
        'period':'na',
        'neg':{'man'},
        'pos':{'woman'},
        'neither':{}
    },{
        'contrast':'Woman-Man',
        'source':'M',
        'period':'na',
        'neg':{'man','boy','brother','father','husband','son'},
        'pos':{'woman','girl','sister','mother','wife','daughter'},
        'neither':{}
    },{
        'contrast':'Good-Bad',
        'source':'S',
        'period':'na',
        'pos':{'good'},
        'neg':{'bad'},
        'neither':{}
    },{
        'contrast':'Future-Past',
        'source':'M',
        'period':'na',
        'pos':{'future','goal','ideal'},
        'neg':{'past','tradition'},
        'neither':{}
    },{
        'contrast':'Future-Past',
        'source':'S',
        'period':'na',
        'pos':{'future'},
        'neg':{'past'},
        'neither':{}
    },
    {
        'contrast':'Future-Past',
        'source':'S',
        'period':'na',
        'pos':{'future'},
        'neg':{'past'},
        'neither':{}
    },{
        'contrast':'Pos-Neg',
        'source':'HGI',
        'period':'na',
        'pos':fieldd['HGI.Pstv'],
        'neg':fieldd['HGI.Ngtv'],
        'neither':{}
    },
    ]
```

```python
def get_key_contrasts():
    origc=[cd for cd in get_origcontrasts() if cd['source']=='Median']
    miscc=get_misccontrasts()
    return origc + miscc
```

```python
pd.DataFrame(get_key_contrasts())
```

```python
def compute_key_vector_scores(m,fieldd=None,words=None):
    # get fields
    if fieldd is None: fieldd=get_fields()
    abswordseed=fieldd['Abs-Conc.Median.Abs.orig']
    concwordseed=fieldd['Abs-Conc.Median.Conc.orig']
    odf=pd.DataFrame()
    for cd in get_key_contrasts():
        odf[cd['contrast']+'.'+cd['source']] = compute_vector_scores(m, cd['pos'], cd['neg'])
    odf=odf.rename_axis('word').reset_index()
    if words: odf=odf[odf.word.isin(set(words))]
    return odf.set_index('word').sort_values('Pos-Neg.HGI')
```

```python
def compute_key_vector_scores_across_models(pathdf,words=None,num_proc=1,**attrs):
    #gby=[c for c in pathdf.columns if not c.startswith('path')]
    gby=['corpus','period','run']
    return pmap_groups(
        do_compute_all_vectors,
        pathdf.groupby(gby),
        num_proc=num_proc,
        kwargs=dict(words=words),
        desc='Computing key vectors across model runs'
    )
def do_compute_all_vectors(df,words=None):
    m=load_model_row(df.iloc[0])
    return compute_key_vector_scores(m,words=words).reset_index()
```

```python
odf=compute_key_vector_scores_across_models(dfmodelsdec, words=words, num_proc=4)
odf
```
